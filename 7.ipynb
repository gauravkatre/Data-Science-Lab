```python
#Perform descriptive statistical analysis (mean, median, std, variance, skewness, kurtosis).
#to derive statistical insights from raw data.
import numpy as np
import pandas as pd
from scipy.stats import skew, kurtosis
```

```python
np.random.seed(42)
```

```python
data = np.random.randint(60, 101, size=30)
```

```python
scores = pd.Series(data)
```

```python
mean_val = scores.mean()
median_val = scores.median()
std_val = scores.std()
var_val = scores.var()
skew_val = skew(scores)
kurtosis_val = kurtosis(scores)
```

```python
print("Randomly Generated Data:", data.tolist())
print("\nDescriptive Statistics:")
print(f"Mean: {mean_val:.2f}")
print(f"Median: {median_val:.2f}")
print(f"Standard Deviation: {std_val:.2f}")
print(f"Variance: {var_val:.2f}")
print(f"Skewness: {skew_val:.2f}")
print(f"Kurtosis: {kurtosis_val + 3:.2f} (Total), {kurtosis_val:.2f} (Excess)")
```
